Importing packages

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

Reading and displaying image

```python
img=cv2.imread(r'D:\Downloads\target image.jpg')
plt.imshow(img)
```

Printing texts from the image

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img))
```

Locating words on image and drawing bounding box around them

```python
hImg,wImg,c=img.shape
boxes=pytesseract.image_to_data(img)

for x,b in enumerate(boxes.splitlines()):
    if x!=0:
        b=b.split()
        print(b)
        if len(b)==12:
            x,y,w,h=int(b[6]),int(b[7]),int(b[8]),int(b[9])
            cv2.rectangle(img,(x,y),(w+x,h+y),(0,0,255),3) 
            cv2.putText(img,b[11],(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(50,50,255),1)
```

Displaying the result

```python
plt.imshow(img)
```
